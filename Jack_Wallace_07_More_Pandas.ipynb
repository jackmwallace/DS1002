```python
# Throws an error after the first run because the string has already been replaced
# Gets rid of the non-numerical portion of the strings
df['duration'] = df['duration'].str.strip('minutes')
# Convert the column 'your_column' to int64
df['duration'] = df['duration'].astype('int64')
print(df.head())
```

```python
# Now display the datatypes in the updated dataframe
print(df.dtypes)
```

```python
# Show the min and max values of ride durations (the shortest and longest rides)
# Hint: If 'df' is the name of your dataframe, simply specify the column name
# and then append the .min() or .max() methods.
print("Minimum ride duration: ", df['duration'].min(), "\nMaximum ride duration: ", df['duration'].max())
```

```python
# Create a new dataframe object and populate it with rides that are 60 minutes or less.
# Hint: Use the .query() method for the dataframe that we used in class.
ndf = df.query('duration <= 60')
```

```python
# Now show both the min and max values of the duration column for your new dataset.
print("Minimum ride duration: ", ndf['duration'].min(), "\nMaximum ride duration: ", ndf['duration'].max())
```

```python
# Using the new dataframe, update it so that is ONLY has records where
# the tire_size is 26.
ndf = ndf[ndf['tire_size'] == 26]
```

```python
# Finally, produce a count of all records with FEMALE riders
# using the 26-inch wheeled rental bikes for their rentals of
# one hour or less.
ndf = ndf[ndf['user_gender'] == "Female"]
print(ndf.head())
```

2. Find the file `more-messy-data.csv` on GitHub and upload it into Google Colab.

Create a new Pandas DataFrame from that data and perform the following steps.

```python
# Write a snippet of code that will display the number of duplicate rows exist in the DF.
from google.colab import files
uploaded = files.upload()
import re
# Load the CSV as a pandas DataFrame
mdf = pd.read_csv('more-messy-data.csv')
```

```python
# Next, remove all duplicate rows using Python.
mdf = mdf.drop_duplicates()
print(mdf.dtypes)
```

```python
# For the FOUR columns involving sepal and petal length and width, update all NaN
# values with the column mean.
mdf_columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
# Turn petal_width into a float so the NaN replacement works
mdf['petal_width'] = mdf['petal_width'].replace({'[^0-9.]': ''}, regex=True).astype(float)
print(mdf.dtypes)
# Replace NaN values with the mean value for the specified columns
for col in mdf_columns:
  mdf[col] = mdf[col].fillna(mdf[col].mean())
mdf
```

```python
# Find any values in the dataframe that are still missing, and delete them.
mdf = mdf.dropna()
```

```python
# How many different species are there?
# Calculate using the function nunique
unique = mdf['species'].nunique()


# Display the number of unique strings in the species column
print("Number of unique speices:", unique)
unique_strings = mdf['species'].unique()
print(unique_strings)
```

```python
# Clean up the 'species' variable so that there are exactly three different values.
mapping = {'virginica' : "virginica", "setosa" : "setosa", "setoso" : "setosa", 'virginia' : "virginica", "setosan" : "setosa", "versicolor" : "versicolor"}
mdf['species'] = mdf['species'].str.lower().replace(mapping)
```
