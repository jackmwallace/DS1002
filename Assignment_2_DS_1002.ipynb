```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import requests

def filter_image_red(image_array):
    """
    Filter the image to retain only the red channel.

    Parameters:
    - image_array: NumPy array representing the original image.

    Returns:
    - red_image: NumPy array representing the image with only the red channel.
    """
    red_image = image_array.copy()
    red_image[:, :, (1,2)] = 0  # Set green and blue channel to 0
    return red_image

def filter_image_green(image_array):
    green_image = image_array.copy()
    green_image[:, :, (0,2)] = 0  # Set red and blue channel to 0
    return green_image

def filter_image_blue(image_array):
    blue_image = image_array.copy()
    blue_image[:, :, (0,1)] = 0  # Set red and green channel to 0
    return blue_image

# Prompt the user to enter the image URL
image_url = input("Enter the image URL: ")

# Fetch the image from the URL
response = requests.get(image_url)

# Open the image using PIL
image = Image.open(requests.get(image_url, stream=True).raw)

# Convert the image to a NumPy array
image_array = np.array(image)

# Filter the image for red, green, and blue channels individually
red_image = filter_image_red(image_array)
green_image = filter_image_green(image_array)
blue_image = filter_image_blue(image_array)

# Filter the original images with no red, green, and blue channels
no_blue_image = image_array.copy()
no_blue_image[:, :, 2] = 0  # Set blue channel to 0

no_green_image = image_array.copy()
no_green_image[:, :, 1] = 0  # Set green channel to 0

no_red_image = image_array.copy()
no_red_image[:, :, 0] = 0  # Set red channel to 0

# Display the images
plt.figure(figsize=(12, 12))

# Original image input
plt.subplot(3, 3, 1)
plt.imshow(image_array)
plt.title('Original Image')
plt.axis('off')

# Three RGB function filtered images
plt.subplot(3, 3, 2)
plt.imshow(red_image)
plt.title('Red Channel Only')
plt.axis('off')

plt.subplot(3, 3, 3)
plt.imshow(green_image)
plt.title('Green Channel Only')
plt.axis('off')

plt.subplot(3, 3, 4)
plt.imshow(blue_image)
plt.title('Blue Channel Only')
plt.axis('off')

# Three "no" images
plt.subplot(3, 3, 5)
plt.imshow(no_red_image)
plt.title('No Red Channel')
plt.axis('off')

plt.subplot(3, 3, 6)
plt.imshow(no_green_image)
plt.title('No Green Channel')
plt.axis('off')

plt.subplot(3, 3, 7)
plt.imshow(no_blue_image)
plt.title('No Blue Channel')
plt.axis('off')

# Hide the empty subplot at the bottom right
plt.subplot(3, 3, 9).axis('off')

plt.tight_layout()
plt.show()

# Reconstruct the combined image using original image channels
combined_image = np.zeros_like(image_array)
combined_image[:, :, 0] = red_image[:, :, 0]
combined_image[:, :, 1] = green_image[:, :, 1]
combined_image[:, :, 2] = blue_image[:, :, 2]

original_image_array = np.array(image)

# Compare the combined image with the original image
comparison_result = np.array_equal(combined_image, original_image_array)

if comparison_result:
    print("\n The combined image is the same as the original image, "+str(comparison_result))
else:
    print("\n The combined image is different from the original image, "+str(comparison_result))
```
